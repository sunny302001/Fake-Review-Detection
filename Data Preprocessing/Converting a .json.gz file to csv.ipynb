Converting a .json.gz file to a csv file

```python
def convert_gz_to_csv(input_folder, output_folder):
    # Check if the output folder exists, if not, create it
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get a list of all files in the input folder
    file_list = os.listdir(input_folder)

    # Process each .gz file and convert to .csv
    for filename in file_list:
        if filename != 'All_Beauty.json.gz' and filename != 'AMAZON_FASHION.json.gz':
            if filename.endswith('.gz'):
                data = []
                input_file_path = os.path.join(input_folder, filename)
                output_file_path = os.path.join(output_folder, os.path.splitext(filename)[0] + '.csv')
                with gzip.open(input_file_path,'rt') as f:
                    for l in f:
                        data.append(json.loads(l.strip()))
                df = pd.DataFrame.from_dict(data)
                df.to_csv(output_file_path)
                df = pd.read_csv(output_file_path)
                print(df.info())
                df.rename(columns={'overall' : 'rating'},inplace = True)
                df.rename(columns = {'Unnamed: 0':'S.No'},inplace = True)
                df.drop(['S.No', 'vote','style','image','asin'], axis=1, inplace=True)
                print(df.info())
                df.to_csv(input_file_path)
                
            
            
```
